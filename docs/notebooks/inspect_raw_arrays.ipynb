```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_arrays_path = "/Users/hayden/Desktop/TDYNO/tdynoDAE-data/dummyset/"
if not os.path.exists(raw_arrays_path):
    print("Error: path does not exist")
```

```python
array_list = [f for f in os.listdir(raw_arrays_path) if f.endswith(".npy")]
```

```python
array_path = os.path.join(raw_arrays_path, array_list[0])
arr = np.load(array_path)

arr[0].shape
arr[1].shape
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(16,9))

arr0 = arr[0]
arr1 = arr[1]
corr_arr = np.subtract(arr[1], arr[0])

axs[0].imshow(arr0)
axs[0].set_title(f"Max pixel intensity: {np.nanmax(arr0)}")
axs[1].imshow(arr1)
axs[1].set_title(f"Max pixel intensity: {np.nanmax(arr0)}")
axs[2].imshow(corr_arr)
axs[2].set_title(f"Max pixel intensity: {np.nanmax(corr_arr)}")
plt.show()
```

```python
arr0 = arr0.astype(np.int32)
arr1 = arr1.astype(np.int32)

one_minus_zero = np.subtract(arr1, arr0)
zero_minus_one = np.subtract(arr0, arr1)

negatives_in_1_minus_0 = len(one_minus_zero[one_minus_zero < 0])
negatives_in_0_minus_1 = len(zero_minus_one[zero_minus_one < 0])

print(negatives_in_0_minus_1 > negatives_in_1_minus_0)
```
