```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
identity_104477 = np.load('/Users/hayden/Desktop/TDYNOdae/results/0.npy')
thresh = 0.46
```

```python
plt.imshow(identity_104477)
plt.title('Denoised image of 104477 with identity filter')
plt.show()
```

```python
#Select only the first 300 columns where the edges are more pronounced
identity_104477 = identity_104477[:, :300]

#Sum along the rows and normalise to row length
rowsum = np.sum(identity_104477, axis=1)
rowsum /= rowsum.shape[0]
rowsum /= np.max(rowsum)
```

```python
x = np.arange(1, 2065)
plt.plot(x, rowsum)
plt.hlines(thresh, xmin=1, xmax=2064, linestyles='-', colors='r')
plt.show()
```

```python
for i, entry in enumerate(rowsum):
    if entry <= thresh:
        rowsum[i] = 1
    else:
        rowsum[i] = 0
```

```python
plt.plot(x, rowsum)
plt.show()
```
